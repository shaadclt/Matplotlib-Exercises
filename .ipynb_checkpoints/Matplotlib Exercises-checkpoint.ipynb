# Axes Labels

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Data")

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

# Legend

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Data")
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend()
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Data")
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(loc='lower left',shadow=True,fontsize='small')
```

# Grid 

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Data")
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend()
plt.grid()
```

# Bar Plot

```python
import numpy as np

company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.ylabel('Revenue in USD Billions')
plt.title("US Tech Companies")
plt.bar(xpos,revenue)
```

## Plotting two values on same bar plot

```python
profit = [40,2,34,12]

xpos = np.arange(len(company))
plt.xticks(xpos,company)
plt.ylabel('Revenue in USD Billions')
plt.title("US Tech Companies")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

## Horizontal Bar Plot

```python
ypos = np.arange(len(company))
plt.yticks(ypos,company)
plt.xlabel('Revenue in USD Billions')
plt.title("US Tech Companies")
plt.barh(ypos-0.2,revenue,height=0.4,label="Revenue")
plt.barh(ypos+0.2,profit,height=0.4,label="Profit")
plt.legend()
```

# Histogram 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.xlabel("Blood Sugar Levels")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='blue')
```

## Plotting Two Values on same Histogram

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("Blood Sugar Levels")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['Men','Women'])
plt.legend()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,400,250]
exp_labels = ['Home Rent','Food','Phone','Car','Other Utilities']
plt.pie(exp_vals,labels = exp_labels)
plt.show
```

```python
plt.pie(exp_vals,labels = exp_labels,radius=1.5) # Setting the pie-chart radius
plt.show
```

## Displaying percentage

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%d%%',shadow=True)
plt.show
```

## Separating a pie  

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%d%%',shadow=True,explode=[0,0,0.3,0,0])
plt.show
```

# Saving the pie-chart to a file

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%d%%',shadow=True,explode=[0,0,0.3,0,0])
plt.show
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=2)
```
